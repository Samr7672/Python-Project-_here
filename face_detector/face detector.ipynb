```python
import cv2
```

```python
face_cap=cv2.CascadeClassifier(r"C:\Users\rebik\OneDrive\Desktop\sample_project_1\env\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")


video_detect=cv2.VideoCapture(0)
while True:
    
    ret, frame = video_detect.read()
    col=cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces=face_cap.detectMultiScale(
        col,
        scaleFactor=1.1,
        minNeighbors=5,
        minSize=(30,30),
        flags=cv2.CASCADE_SCALE_IMAGE
    )
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),2)
    if not ret:
        print("Failed to grab frame")
        break

    
    cv2.imshow("video_live", frame)
    if cv2.waitKey(15) == ord("a"):
        break
video_detect.release()
```
